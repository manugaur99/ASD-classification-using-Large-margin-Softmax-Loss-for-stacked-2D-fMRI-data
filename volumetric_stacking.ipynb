Selecting images with threshold pixels,stacking them and saving as



Dataset (1 hour to run)

```python
#ccs 2d has all slices for 3d image file for each patient

dataset = '/content/drive/MyDrive/ABIDE_NYU/more_ccs_2D/'
# iterate all the 174 patients 
for foldername in os.listdir(dataset):
    folder = os.path.join(dataset, foldername)

    directory = folder + '/' 
    # iterate all the slices for a patient
    threshold = 30
    img_list = []
    for filename in os.listdir(directory):
        f = os.path.join(directory, filename)
        # checking if it is a file
        if os.path.isfile(f):
          img = cv2.imread(f)
          if(img.mean()>30):
            img_list.append(img)      #len = 46

    #Let's stack these slices 
    #one channel has only 1 channel of all images for a patient
    onechannel = []
    for img in img_list:
      onechannel.append(img[:,:,0])
    stacked_image = cv2.merge(onechannel)    # shape >> 73,61,46
    #save this stacked image 
    np.save('/content/drive/MyDrive/ABIDE_NYU/more_Dataset/' + foldername,stacked_image)
```
